```python
import sys
sys.path.append('PycharmProjects/mlia/mliac6')
import svm
dataArr, labelArr = svm.loadDataSet('testSet.txt')
```

```python
from numpy import *
dataMatrix = mat(dataArr)
labelMat = mat(labelArr).transpose()
m, n = shape(dataMatrix)
b = 0
alphas = mat(zeros((m, 1)))
alphas
```

```python
b, alphas = svm.smoP(dataArr, labelArr, 0.6, 0.001, 40)
print("b: %s" % b)
print("alphas(>0): %s" % alphas[alphas>0])
```

```python
ws = svm.calcWs(alphas, dataArr, labelArr)
ws[1]
```

```python
import matplotlib.pyplot as plt
dataArr = array(dataMatrix)
n = shape(dataArr)[0]
xcord1 = []
ycord1 = []
xcord2 = []
ycord2 = []
xcord3 = []
ycord3 = []
xcord4 = []
ycord4 = []
for i in range(n):
    if int(labelMat[i]) == 1:
        if alphas[i] > 0.0:
            xcord3.append(dataArr[i, 0])
            ycord3.append(dataArr[i, 1])
        else:
            xcord1.append(dataArr[i, 0])
            ycord1.append(dataArr[i, 1])
    else:
        if alphas[i] > 0.0:
            xcord4.append(dataArr[i, 0])
            ycord4.append(dataArr[i, 1])
        else:
            xcord2.append(dataArr[i, 0])
            ycord2.append(dataArr[i, 1])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(xcord1, ycord1, s=30, c='red', marker='s')
ax.scatter(xcord2, ycord2, s=30, c='green')
ax.scatter(xcord3, ycord3, s=30, c='red', marker='+')
ax.scatter(xcord4, ycord4, s=30, c='green', marker='+')
x = arange(-1.0, 10.0, 0.1)
y = ((- ws[0] * x - b) / ws[1]).transpose()
ax.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
